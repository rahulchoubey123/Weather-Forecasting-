```python
from tkinter import *
#Importing JSON Module
import json
#Importing request module
import requests
mainwin=Tk() 
mainwin.title("Weather forecast")
mainwin.iconbitmap("770b805d5c99c7931366c2e84e88f251.ico")
bg = PhotoImage(file="301566_beautiful-hd-wallpapers-weather-and-phenomena-backgrounds_1920x1080_h.png")
my_label = Label(mainwin,image=bg)
my_label.place(x=1, y=1, relwidth=1, relheight=1)
labl_city=Label(mainwin,text="City",font=100,bg="Yellow").place(x=650,y=250)
city=StringVar()
entrybox=Entry(mainwin,textvariable=city,width=50)
entrybox.place(x=700,y=250)
#Sending request to open weather appliocation
def forecast():
    response=requests.get("http://api.openweathermap.org/data/2.5/weather?"+"appid="+"dae017122c460176252eaaf4a41da0bf"+"&q="+entrybox.get())
#Converting json data in to python dictionary
    weatherdata=response.json()
    #print(weatherdata)
    #print("Weather analysis for",city)
    #Finding Coordinates of city
    Coordinates=weatherdata['coord']
    Latitude=Coordinates['lat']
    lbl_lat.configure(text=Latitude)
    Longitude=Coordinates['lon']
    lbl_lon.configure(text=Longitude)
#Finding Temperature of City
    Temperature=weatherdata['main']
    temp=Temperature['temp']
    lbl_temp.configure(text=temp)
    Feelslike=Temperature['feels_like']
    lbl_feels.configure(text=Feelslike)
    tempmin=Temperature['temp_min']
    lbl_min.configure(text=tempmin)
    tempmax=Temperature['temp_max']
    lbl_max.configure(text=tempmax)
    pressure=Temperature['pressure']
    lbl_press.configure(text=pressure)
    humidity=Temperature['humidity']
    lbl_hum.configure(text=humidity)
    vis=weatherdata['visibility']
    lbl_vis.configure(text=vis)
#Wind
    wind=weatherdata['wind']
    speed=wind['speed']
    lbl_speed.configure(text=speed)
    weather=weatherdata['weather']
    lis=weather[0]
    description=lis['description']
    lbl_des.configure(text=description)

button=Button(mainwin,text="Search",command=forecast,height=2).place(x=830,y=270)
lbl1=Label(mainwin,text="Latitude:",font=80,bg="Purple",fg="White").place(x=450,y=370)
lbl2=Label(mainwin,text="Longitude:",font=80,bg="Purple",fg="White").place(x=1050,y=370)
lbl3=Label(mainwin,text="Temperature:",font=80,bg="Purple",fg="White").place(x=435,y=450)
lbl4=Label(mainwin,text="Feels like:",font=80,bg="Purple",fg="White").place(x=1050,y=450)
lbl5=Label(mainwin,text="Min Tempearture:",font=80,bg="Purple",fg="White").place(x=420,y=530)
lbl6=Label(mainwin,text="Max Temperature:",font=80,bg="Purple",fg="White").place(x=1025,y=530)
lbl7=Label(mainwin,text="Pressure:",font=80,bg="Purple",fg="White").place(x=450,y=610)
lbl8=Label(mainwin,text="Humidity:",font=80,bg="Purple",fg="White").place(x=1050,y=610)
lbl9=Label(mainwin,text="Visibility:",font=80,bg="Purple",fg="White").place(x=450,y=690)
lbl10=Label(mainwin,text="Wind Speed:",font=80,bg="Purple",fg="White").place(x=1045,y=690)
lbl1=Label(mainwin,text="Description:",font=80,bg="Purple",fg="White").place(x=670,y=750)

lbl_lat=Label(mainwin,text="----",font=80,bg="Orange")
lbl_lat.place(x=600,y=370)
lbl_lon=Label(mainwin,text="----",font=80,bg="Orange")
lbl_lon.place(x=1200,y=370)
lbl_temp=Label(mainwin,text="----",font=80,bg="Orange")
lbl_temp.place(x=600,y=450)
lbl_feels=Label(mainwin,text="----",font=80,bg="Orange")
lbl_feels.place(x=1200,y=450)
lbl_min=Label(mainwin,text="----",font=80,bg="Orange")
lbl_min.place(x=600,y=530)
lbl_max=Label(mainwin,text="----",font=80,bg="Orange")
lbl_max.place(x=1200,y=530)
lbl_press=Label(mainwin,text="----",font=80,bg="Orange")
lbl_press.place(x=600,y=610)
lbl_hum=Label(mainwin,text="----",font=80,bg="Orange")
lbl_hum.place(x=1200,y=610)
lbl_vis=Label(mainwin,text="----",font=80,bg="Orange")
lbl_vis.place(x=600,y=690)
lbl_speed=Label(mainwin,text="----",font=80,bg="Orange")
lbl_speed.place(x=1200,y=690)
lbl_des=Label(mainwin,text="----",font=80,bg="Orange")
lbl_des.place(x=800,y=750)
title=Label(mainwin,text="REAL TIME WEATHER FORECAST",font=150,bg="Orange").place(x=700,y=100)
lbl=Label(mainwin,text="Enter the city name",font=300,bg="Pink").place(x=770,y=210)
mainwin.mainloop()
```
